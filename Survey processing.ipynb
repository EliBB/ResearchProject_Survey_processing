```python
# Linear algebra
import numpy as np

# Data processing and CSV file
import pandas as pd

# Data visualization
import matplotlib.pyplot as plt

# Read data from CSV file
data = 'For the love of knitting.csv'
df = pd.read_csv(data)
```

```python
df.head()
```

```python
col_names = ['Date', 'Experience', 'Frequency', 'Hours', 'Country', 'Reason', 'Introduction', 'Place', 'Inspiration', 'FavoriteKnittingProject', 'SizeProject','ChooseYarn', 'BuyYarn', 'Notes', 'OverallTools','ImportantTools', 'Level', 'Motivation', 'GetHelp', 'PatternType', 'AppTools', 'SharingData', 'WhyShare?', 'AnythingElse', 'test']

df.columns = col_names

df.columns
```

```python
df.head()
```

```python
ExperienceCount = df.Experience.value_counts()

x = ExperienceCount

labels = 'More than 20 years', '11-20 years', '1-5 years', '6-10 years',

colors = ['#FFF5F5', '#E3CDCC', '#C4A9A5', '#795C58']

wedgeprops = { 'edgecolor' : '#666666'}


fig, ax = plt.subplots()
ax.pie(x, 
       autopct='%1.1f%%',
       labels=labels, 
       colors=colors,
       wedgeprops = wedgeprops,
       startangle=15)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Knitting experience')
plt.show()
```

```python
Country = df['Country']

listOfContinents = []

def sortCountries():
    
    listOfContinents = []

    for knitter in Country:
        
        if knitter == 'USA' or knitter == 'Canada':
            Continent = 'North America'
            #listOfContinents.append(Continent)
            
        elif knitter == 'Japan':
            Continent = 'Asia'
            #listOfContinents.append(Continent)
            
        elif knitter == 'New Zealand':
            Continent = 'Australia' 
            #listOfContinents.append(Continent)
            
        else:
            Continent = 'Europe'
            #listOfContinents.append(Continent)
            
        listOfContinents.append(Continent)
    
    
    #print(listOfContinents)
    print(pd.value_counts(np.array(listOfContinents)))
    #return listOfContinents

sortCountries()



#North America    115
#Europe            44
#Asia               1
#Australia          1


data = [115, 44, 1, 1]



labels = 'North Amercia', 'Europe', 'Asia', 'Australia'

colors = ['#FFF5F5', '#E3CDCC', '#C4A9A5', '#795C58']

wedgeprops = { 'edgecolor' : '#666666'}


fig, ax = plt.subplots()
ax.pie(data, 
       autopct='%1.1f%%',
       labels=labels, 
       colors=colors,

       wedgeprops = wedgeprops,

       startangle=15)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Geographical location')
plt.show()


```
